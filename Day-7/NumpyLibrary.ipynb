```python
from re import A
#NumPy Array Join
a_one=np.array([10,2,3,41])
a_two=np.array([5,16,17,8])
array_ab=np.concatenate((a_one,a_two))
print(array_ab)

```

```python
#Numpy Array Split
array_split=np.array_split(array_ab,4)
print(array_split)
```

```python
array_split3=np.array_split(array_ab,3)
print(array_split3)
```

```python
#Sorting the NumpyArray
array_sort=np.sort(array_ab)
print(array_sort)
```

```python
#Array of All items less than 20
array_less_than_20=array_ab[array_ab<20]
print(array_less_than_20)
```

```python
array_not_10=array_ab[array_ab!=10]
print(array_not_10)
```

```python
student_marks=np.array(
    [
        [78,80,86,90,85,91],
        [90,71,86,90,75,90],
        [98,88,80,90,83,88],
        [88,83,80,90,82,76],
        [67,89,81,90,71,73],
    ]

)
# * Here Each Row =One student
# *  Each Column =marks in subject
# *  we have six subjects and five students
```


#Exercise
1. Find out dimension,size, and datatype of array
2. Find out aver marks of each student
3. Find out highest marks  in each subject
4. Find out overall class average
5. Find out index(row,colum) of the lowest mark in entire array

```python
#Here Each Row =One student
# Each Column =marks in subject
# we have six subjects and five students
#Exercise
```

```python
#Answer 1
print(' Dimension',student_marks.ndim)
print(' Shape',student_marks.shape)
print(' size',student_marks.size)
print(' type',student_marks.dtype)
```

```python
#Find out aver marks of each student
average_marks=student_marks.mean(axis=1)
print(average_marks)
#
```

```python
#Find out highest marks in each subject
highest_marks=student_marks.max(axis=0)
print(highest_marks)
#
```

```python
#Find out overall class average
overall_class_average=student_marks.mean()
print(overall_class_average)
#
```

```python
#Find out index(row,colum) of the lowest mark in entire array
total_marks=student_marks.sum(axis=1)
print(total_marks)
top_student_index=np.argmax(total_marks)
print(f'Top student index {top_student_index}')
#
#
```
