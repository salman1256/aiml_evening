```python
a_one=np.array([10,2,3,4,15])
a_two=np.array([5,12,16,48,5])
array_ab=np.concatenate((a_one,a_two))
print(array_ab)
```

```python
a_one=np.array([10,2,3,4,15])
a_two=np.array([5,12,16,48,5])
a_three=np.array([1,2,3,4,5])
array_ab=np.concatenate((a_one,a_two,a_three))
print(array_ab)
```

```python
#Numpy  Array split
array_split=np.array_split(array_ab,3)
print(array_split)
```

```python
sorted_array=np.sort(array_ab)
print(sorted_array)
```

```python
# Create Array of All Items less than 10
array_less_than_10=array_ab[array_ab<10]
print(array_less_than_10)
```

```python
# Create Array of All Items less than 10
array_not_5=array_ab[array_ab!=5]
print(array_not_5)
```

```python
#create a numPy Array for following numbers:10,12,13,45,68,69,70,89
#find out all numbers using filter
```

```python
given_array=np.array([10,12,13,45,68,69,70,89])
```

```python
print(given_array)
```

```python
odd_numbers=given_array[given_array%2!=0]
print(odd_numbers)
```

```python
even_numbers=given_array[given_array%2==0]
print(even_numbers)
```
