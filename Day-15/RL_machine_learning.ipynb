```python
import numpy as np
import random
```

```python
states=[0,1,2] # 0=low , 1=medium , 2=high
actions=[0,1] #off the Pump , Pump On
```

```python
print('states',states)
print('Actions',actions)
```

```python
Q=np.zeros((len(states),len(actions)))
alpha=0.1 #Learning Rate
gamma=0.9 # Discount Factor
epsilon=0.2 #Exploration Rate
episodes=300 #Training Runs
print('Q-Table Shape: ',Q.shape)
```

```python
def get_reward(state):
  if(state==1):
    return 10
  elif(state==0):
    return -10
  else:
    return -10
```

```python
print(get_reward(1))
```

```python
def next_state(state,action):
  if(action==1):
    return min(state+1,2)
  else:
    return max(state-1,0)
```

```python
for episode in range(episodes):
  state=random.choice(states)
  for step in range(10):
    if random.random()<epsilon:
      action=random.choice(actions)
    else:
      action=np.argmax(Q[state])

    new_state=next_state(state,action)
    reward=get_reward(new_state)
    Q[state,action]=Q[state,action]+alpha*(reward+gamma*np.max(Q[new_state])-Q[state,action])
    state=new_state


```

```python
print('Learned Q Table')
print(Q)
```

**Complete Example WaterTank using RL**

```python
import numpy as np
import random
```

```python
states=np.arange(0,101,10)
actions=['Fill','STOP']
print('States: ',states)
print('Actions: ',actions)
```

```python
Q=np.zeros((len(states),len(actions)))
alpha=0.1
gamma=0.9
epsilon=0.2
episodes=300
print('Q-Table Shape: ',Q.shape)
```

```python
def get_reward(level,action):
  if 40<=level<=70:
    reward=10
  else:
    reward=-10
  if action=='Fill' and level>=90:
    reward-=10
  if action=='STOP' and level<=10:
    reward-=10

  return reward
```

```python
print('Reward Example 1: ',get_reward(60,'Fill'))
```

```python
def next_level(level,action):
  if action=='Fill':
    level+=random.choice([5,10,15])
  else:
    level-=random.choice([5,10,15])
  return int(np.clip(level,0,100))
```

```python
for ep in range(episodes):
  level=random.choice(states)
  for _ in range(15):
    if random.uniform(0,1)<epsilon:
      action=random.choice(actions)
    else:
      action=actions[np.argmax(Q[level//10])]
    next_state=next_level(level,action)
    reward=get_reward(next_state,action)
    a=actions.index(action)
    best_next=np.max(Q[next_state//10])
    Q[level//10,a]+=alpha*(reward+gamma*best_next-Q[level//10,a])
print ('Training Completed')
```

```python
try:
  level=int(input('Enter Level(0-100): '))
  if level<0 or level>100:
    raise ValueError('Water level out of range')
except ValueError as ve:
  print(ve)
  level=50
  print('Default level set to 50%')
for step in range(10):
  action=actions[np.argmax(Q[level//10])]
  print('Action: ',action)
  level=next_level(level,action)
  print('Level: ',level)

print('\n Simulation Complete . Water Tank Control Done')

```
