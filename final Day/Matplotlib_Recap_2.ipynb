```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_netflix=pd.read_csv('/content/sample_data/netflix_data.csv')
df_netflix.head()
```

```python
type_counts=df_netflix['type'].value_counts()
type_counts
```

```python
plt.figure(figsize=(6,6))
plt.bar(type_counts.index,type_counts.values,color=['red','green'])
plt.xlabel('Type',color='blue')
plt.ylabel('Count',color='blue')
plt.xticks(rotation=0,color='blue')
plt.yticks(color='blue')
plt.title('Distribution of Movie and TV Shows',color='blue')
plt.grid(linestyle='--',color='blue')
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.pie(type_counts.values,labels=type_counts.index, autopct='%1.1f%%',colors=['orange','green'],explode=(0.2,0),startangle=90)
plt.title('Distribution of Movie and TV Shows',color='blue')
plt.legend(title='Type',loc='upper right')
plt.tight_layout()
plt.show()
```
