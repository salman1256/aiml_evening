### *Matplotlib* is a powerful python library used for data visualization. It is used in Data Science, Machine learning , Analytics and research.
You can create different kind of charts using matplot lib.
e.g.
* Line Charts
* Bar Charts
 * Scatter Plots
 * Histograms
 * Pie Charts etc.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep']
sales=[200,250,300,280,350,400,450,500,550]
```

```python
plt.plot(months,sales,marker='o')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Report')
plt.grid(True)
plt.show()
```

```python
df_product_sales=pd.DataFrame(
    {
        'Product':['Laptop','Mobile','Touch Screen','Smart Watch','Tablet'],
        'Sales':[1000,1500,800,1200,2000]
    }
)
```

```python
#Bar chart based on df_product_sales
plt.bar(df_product_sales['Product'],df_product_sales['Sales'])
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Scatter Plots
customer_age=np.random.randint(13,70,30)
customer_salary=np.random.randint(20000,100000,30)
plt.scatter(customer_age,customer_salary)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Customer Age vs Salary')
plt.grid(True)
plt.show()
```

```python
#pie plot example
df_product_sales=pd.DataFrame(
    {
        'Product':['Laptop','Mobile','Touch Screen','Smart Watch','Tablet'],
        'Sales':[1000,1500,800,1200,2000]
    }
)
plt.pie(df_product_sales['Sales'],labels=df_product_sales['Product'])
plt.title('Product Sales')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    df_product_sales['Sales'],
    labels=df_product_sales['Product'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Product Sales')
plt.show()
#
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    df_product_sales['Sales'],
    labels=df_product_sales['Product'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['red','green','blue','yellow','orange'],
    explode=[0.1,0,0,0,0]
)
plt.title('Product Sales')
plt.show()
```

```python
# sub plot example
months=['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep']
sales=[200,250,300,280,350,400,450,500,550]

df_product_sales=pd.DataFrame(
    {
        'Product':['Laptop','Mobile','Touch Screen','Smart Watch','Tablet'],
        'Sales':[1000,1500,800,1200,2000]
    }
)

customer_age=np.random.randint(13,70,30)
customer_salary=np.random.randint(20000,100000,30)

order_values=np.random.normal(500,150,100)
```

```python
order_values
```

```python
fig, ax= plt.subplots(2,2,figsize=(12,8))
# ax[0,0] line chart
ax[0,0].plot(months,sales,color='green',marker='o',linewidth=3)
ax[0,0].set_title('Sales per Month',color='green')
ax[0,0].set_xlabel('Months',color='green')
ax[0,0].set_ylabel('Sales',color='green')
ax[0,0].tick_params(axis='x',colors='green')
ax[0,0].tick_params(axis='y',colors='green')
ax[0,0].grid(color='lightgreen',linestyle='--')

# 2 bar chart ax[0,1]

bar_colors=['red','green','blue','yellow','orange']
ax[0,1].bar(df_product_sales['Product'],df_product_sales['Sales'],color=bar_colors)
ax[0,1].set_title('Product Sales',color='blue')
ax[0,1].set_xlabel('Product',color='blue')
ax[0,1].set_ylabel('Sales',color='blue')
ax[0,1].tick_params(axis='x',colors='blue')
ax[0,1].tick_params(axis='y',colors='blue')
ax[0,1].grid(color='lightblue',linestyle='--')
plt.tight_layout()

# 3 chart : ax[1,0] , scatter plot
ax[1,0].scatter(customer_age,customer_salary,color='orange')
ax[1,0].set_title('Customer Age vs Salary',color='orange')
ax[1,0].set_xlabel('Age',color='orange')
ax[1,0].set_ylabel('Salary',color='orange')
ax[1,0].tick_params(axis='x',colors='orange')
ax[1,0].tick_params(axis='y',colors='orange')
ax[1,0].grid(color='brown',linestyle='--')

# 4 chart : ax[1,1], histogram
ax[1,1].hist(order_values,bins=20,color='purple')
ax[1,1].set_title('Order Values Distribution',color='purple')
ax[1,1].set_xlabel('Order Values',color='purple')
ax[1,1].set_ylabel('Frequency',color='purple')
ax[1,1].tick_params(axis='x',colors='purple')
ax[1,1].tick_params(axis='y',colors='purple')
ax[1,1].grid(color='pink',linestyle='--')
plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(6,4))
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend()
plt.show()
```
