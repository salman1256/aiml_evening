```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
months=['Jan','Feb','Mar','April']
sales=[20000,30000,25000,36000]
```

```python
plt.figure(figsize=(6,3))
plt.plot(months,sales,linestyle='--',linewidth=2,marker='*',label='monthly sales',color='red')
plt.xlabel('months')
plt.ylabel('sales')
plt.title('monthly sales')
plt.legend()
plt.xlim(0,4)
plt.ylim(20000,40000)
plt.legend(loc='lower right')
plt.tight_layout=()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
bar_colors=['red','green','blue','yellow']
plt.bar(months,sales,color=bar_colors,label='monthly sales')
plt.xlabel('months')
plt.ylabel('sales')
plt.title('monthly sales')
plt.legend()
plt.xlim(0,4)
plt.ylim(20000,40000)
plt.legend(loc='upper left')
plt.tight_layout=()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X-Lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend()
plt.show()
```

**Histogram:** Used for data visualization to repersent of data distrubution showing how frequent each value in given range occurs.

```python
data=np.random.randn(1000)
print(data)
```

```python
plt.figure(figsize=(6,4))
plt.hist(data,bins=30,color='magenta',edgecolor='black',alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('First Histogram')
plt.show()
```

* create  histogram based on Netflix data released Movies adn TV-Shows by Year


```python
df_netflix=pd.read_csv('/content/sample_data/netflix_data.csv')
```

```python
df_netflix.head()
```

```python
movies=df_netflix[df_netflix['type']=='Movie']
print(movies)
```

```python
shows=df_netflix[df_netflix['type']=='TV Show']
print(shows)
movies['release_year']
shows['release_year']
```

```python
plt.figure(figsize=(6,4))
plt.hist(movies['release_year'],bins=30,label='Movies',edgecolor='black',color='red')
plt.hist(shows['release_year'],bins=30,label='Shows',edgecolor='black',color='green')
plt.title('Netflix Data')
plt.xlabel('Score')
plt.ylabel('Number of Titles released')
plt.legend(title='Movies/Shows')
plt.grid(axis='y',linestyle='--',color='skyblue')
plt.savefig('netflix2_hist.jpeg',dpi=300)
plt.show()
```

```python
type_counts=df_netflix['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(6,4))
plt.pie(type_counts,labels=type_counts.index,autopct='%1.2f%%',startangle=140,colors=['green','darkblue'],explode=(0.1,0))
plt.title('Netflix Movies and TV Shows')
plt.legend(title='Type',loc='lower right')
plt.savefig('netflix_pieplot.pdf',dpi=300)
plt.show()
```
