Subplots(): provides a way ot plot multiple plots on a single figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(6,6))
```

```python
our_colors=['red','green','blue','orange']
sales_data={'Electronics':4500,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
sales=list(sales_data.values())
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(8,6))
ax1.bar(category,sales,color=our_colors)
ax1.set_xlabel('Category')
ax1.set_ylabel('Sales')
ax1.set_title('Sales by Category')
x=np.random.randn(1000)

ax2.hist(x,bins=20,color='orange')
ax2.set_title('Histogram')

ax3.plot(x,x/2,color='green')
ax3.set_title('Line Plot')

ax4.scatter(np.random.randn(100),np.random.randn(100),color='red')
ax4.set_title('Scatter Plot')
plt.tight_layout()
plt.show()
```
