**K-means Clustering**
* Cluster: We will partition our dataset into K distinct group (Cluster)
* Centroid: Nearest mean cluster


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset ={
    'Entry_time':[9,10,11,14,15,16,17,18,19,20],
    'Duration':[30,45,40,90,85,100,105,110,125,130]
}
```

```python
df=pd.DataFrame(dataset)
df
```

```python
model_kmeans=KMeans(n_clusters=3,random_state=42) #Create Model
```

```python
df['Cluster']=model_kmeans.fit_predict(df) # Train and predict
```

```python
df
```

```python
centroid=model_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Entry_time'],df['Duration'],c=df['Cluster'])
plt.scatter(centroid[:,0],centroid[:,1],color='red',marker='*')
plt.xlabel('Entry_time')
plt.ylabel('Duration')
plt.title('K-means Clustering')
plt.grid(True,linestyle='--',color='lightblue')
plt.show()
```

**Hierarchical Clustering**: is unsupervised Machine learning technique that groups data into a hierarachiy of clusters.
* Build a tree like structure
* No need to choose the cluster intially
* shows realationship using tree like structure diagram(dendogram)

```python
#import all requied libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
df=pd.DataFrame(
    {
        'Experience':[1,2,3,7,8,9,10,4],
        'Salary':[2500,2800,3000,7000,7500,8000,9000,3500]
    }
)
```

```python
df
```

```python
#Creating dendogram for this dataframe
#create linkage matrix
linkage_matrix=linkage(df,method='ward',metric='euclidean')
```

```python
plt.figure(figsize=(8,4))
dendrogram(linkage_matrix)
plt.title('Dendogram')
plt.xlabel('Employee')
plt.ylabel('Distance')
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
df['Cluster']=model.fit_predict(df)
```

```python
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Experience'],df['Salary'],c=df['Cluster'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Hierarchical Clustering')
plt.show()
```

* Create Dendogram for this dataframe
* Create cluster from Dendogram
* Visualize the Cluster (i.e. Create Scatter Plot)


```python
#Create Dendogram for this dataframe
linkage_matrix=linkage(df,method='ward',metric='euclidean')
```

```python
plt.figure(figsize=(8,4))
dendrogram(linkage_matrix)
plt.title('Dendogram')
plt.xlabel('Customer')
plt.ylabel('Distance')
plt.show()
```

```python
plt.figure(figsize=(8,4))
dendrogram(linkage_matrix,labels=df.index,leaf_rotation=90,distance_sort='descending')
plt.title('Dendogram')
plt.xlabel('Customer')
plt.ylabel('Distance')
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
df['Cluster']=model.fit_predict(df)
```

```python
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['monthly_spend'],df['visit_per_month'],c=df['Cluster'])
for i in df.index:
  plt.text(df.loc[i,'monthly_spend'],df.loc[i,'visit_per_month'],i)

plt.xlabel('monthly_spend')
plt.ylabel('visit_per_month')
plt.title('Customers Hierarchical Clustering')
plt.show()
```

DBSCAN:Desnity Based Spatial Clustering of Application with Noise
* Does not require K
* finds cluster of any shape
* automatically detects noise
* useful in real world messy data
* eps: Maximum distance between two points to be considered as neighbors
* min_samples: Minimum number of points to form a dense region


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
dataset={
    'Latitude':[11.0,12.97,12.98,12.99,13.0,13.01,13.02,13.03,13.04,19.07,19.08,19.09,28.61],
    'Longitude':[75.59,77.60,77.61,77.62,72.87,72.88,72.89,72.90,72.95,72.91,72.95,72.96,77.9]
    }
```

```python
df =pd.DataFrame(dataset)
```

```python
df.head()
```

```python
df.shape
```

```python
# from sklearn.cluster import DBSCAN
# from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
df_scaled=scaler.fit_transform(df)
#
```

```python
model_dbscan=DBSCAN(eps=0.1,min_samples=3,metric='euclidean')
```

```python
df['Cluster']=model_dbscan.fit_predict(df_scaled)
```

```python
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Latitude'],df['Longitude'],c=df['Cluster'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('DBSCAN Clustering')
plt.show()
```
