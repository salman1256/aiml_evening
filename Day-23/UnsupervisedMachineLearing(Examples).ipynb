**Unsupervised Machine Learning**:No labels (no target/output) machine looks at data and finds hidden partterns by itself
* K-means Clustering
* Hierarchical Clustering

**K-means Clustering**
* Cluster: We will partition our dataset into K distinct group (Cluster)
* Centroid: Nearest mean cluster


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dataset ={
    'Entry_time':[9,10,11,14,15,16,17,18,19,20],
    'Duration':[30,45,40,90,85,100,105,110,125,130]
}
```

```python
df=pd.DataFrame(dataset)
df
```

```python
model_kmeans=KMeans(n_clusters=3,random_state=42) #Create Model
```

```python
df['Cluster']=model_kmeans.fit_predict(df) # Train and predict
```

```python
df
```

```python
centroid=model_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Entry_time'],df['Duration'],c=df['Cluster'])
plt.scatter(centroid[:,0],centroid[:,1],color='red',marker='*')
plt.xlabel('Entry_time')
plt.ylabel('Duration')
plt.title('K-means Clustering')
plt.grid(True,linestyle='--',color='lightblue')
plt.show()
```

```python
df=pd.read_csv('/content/sample_data/customers.csv')
```

```python
df.head()
```

```python
df_customers=df[['AnnualIncome',	'SpendingScore']]
```

```python
df_customers
```

```python
model_kmeans=KMeans(n_clusters=5,random_state=42)
```

```python
df['Cluster']=model_kmeans.fit_predict(df_customers)
```

```python
df
```

```python
centroid=model_kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df_customers['AnnualIncome'],df_customers['SpendingScore'],c=df['Cluster'])
plt.scatter(centroid[:,0],centroid[:,1],color='red',marker='D')
plt.xlabel('AnnualIncome')
plt.ylabel('SpendingScore')
plt.title('K-means Clustering')
plt.grid(True,linestyle='--',color='lightblue')
plt.show()
```
