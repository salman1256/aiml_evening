**One Hot Encoding Example**

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
```

```python
data={
    'Color':['Red','Blue','Green','Red','Blue']
    }
```

```python
df=pd.DataFrame(data)
df
```

```python
ohe=OneHotEncoder()
```

```python
encoded=ohe.fit_transform(df[['Color']])
```

```python
df
```

```python
encoded_df=pd.DataFrame(encoded.toarray(),columns=ohe.get_feature_names_out())
encoded_df
```

* OneHot Encoding using panadas get_dummies()

```python
df_students=pd.DataFrame({
    'Student ID':[1,2,3,4,5],
    'Subject':['Math','Science','English','Math','Science']
})
df_students
```

```python
one_hot_encoded=pd.get_dummies(df_students['Subject'])
one_hot_encoded
```

```python
df_students=pd.concat([df_students,one_hot_encoded],axis=1)
df_students
```

```python
data={
    'Player':['Sachin','Moin Ali','S.Broad','Bumrah','Ricky Pointing','Max Well','Dhoni'],
    'Country':['India','England','England','India','Aus','Aus','India']
}
```

```python
df_players=pd.DataFrame(data)
df_players
```

**Apply one hot Encoding  on player's country and display updated data frame**

```python
one_hot_encoded=pd.get_dummies(df_players['Country'],prefix='Country')
one_hot_encoded
```

```python
df_players=pd.concat([df_players,one_hot_encoded],axis=1)
df_players
```
