**SubPlot:**: multiple plots inside a single figure.
* plt.subplot(rows,columns,index)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
our_colors=['red','green','blue','orange']
sales_data={'Electronics':4500,'Clothing':30000,'Grocery':25000,'Mobile':20000}
category=list(sales_data.keys())
sales=list(sales_data.values())
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.set_title('Bar Chart')
ax1.grid(True)
ax2.set_title('Scatter Plot')
ax3.set_title('Pie Chart')
ax4.set_title('Histogram')
ax1.bar(category,sales,color=our_colors)
ax2.scatter(category,sales,color=our_colors)
ax3.pie(sales,labels=category,colors=our_colors)
ax4.hist(sales,bins=5)
plt.tight_layout()
plt.show()
#
```
