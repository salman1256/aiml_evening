**Seaborn Library of Matplotlib**
* Seaborn is a data visualization library in matplotlib.
* Seaborn provides high level interface for creating interactive and more colorful plots using less code.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
palette_our = sns.color_palette() #Return a list of colors or continuous colormap defining a palette.
                              #Possible palette values include


```

```python
sns.palplot(palette_our)
```

```python
palette_green=sns.color_palette("Greens",15)
sns.palplot(palette_green)
```

```python
palette_red=sns.color_palette("Reds",15)
sns.palplot(palette_red)
```

```python
data={
    'buildings':['Burj Khalifa','Medreka','Shangai Tower','Qutub Minar','Al-Bait Clock'],
     'category':['highest','medium','medium','low','medium'],
    'Height_m':[800,678,632,200,601]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(6,4))
#bar_colors=sns.color_palette('Greens',3)
bar_colors=['red','green','yellow']
sns.barplot(x='buildings',y='Height_m',data=df,hue='category',palette=bar_colors)
plt.xlabel('Buildings')
plt.ylabel('Height in meter')
plt.title('Height of Buildings')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
sns.set_style('white') #whitegrid,darkgrid,ticks,white
sns.set_palette('dark')  #pastel,dark, muted,viridis,magma,coloblind,rocket,inferno
plt.figure(figsize=(8,6))
sns.barplot(x='buildings',y='Height_m',data=df,hue='category')
plt.xticks(rotation=45)
plt.xlabel('Buildings')
plt.ylabel('Height in meter')
plt.title('Height of Buildings')
plt.legend(loc='best',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
data_iris=sns.load_dataset('iris')
data_iris.head()
```

```python
sns.set_style('white') #whitegrid,darkgrid,ticks,white
sns.set_palette('inferno')  #pastel,dark, muted,viridis,magma,coloblind,rocket,inferno
plt.figure(figsize=(8,6))
sns.barplot(x='species',y='sepal_width',data=data_iris,hue='species')
plt.title('Barplot of Iris Dataset')
plt.xlabel('Name of Species')
plt.ylabel('Sepal Width in cm')
plt.show()
```

```python
sns.violinplot(x='species',y='sepal_width',data=data_iris,hue='species')
plt.title('Violinplot of Iris Dataset')
plt.xlabel('Name of Species')
plt.ylabel('Sepal Width in cm')
plt.grid(axis='y', linestyle='--',color='lightblue')
plt.show()
```

```python
sns.stripplot(x='species',y='sepal_width',data=data_iris,hue='species')
plt.title('Stripplot of Iris Dataset')
plt.xlabel('Name of Species')
plt.ylabel('Sepal Width in cm')
plt.grid(axis='y', linestyle='--',color='lightblue')
plt.show()
```

```python

data_iris=sns.load_dataset('iris')
sns.pairplot(data_iris,hue='species',palette='Set2')
plt.show()
```

```python
sns.set_style('darkgrid') #whitegrid,darkgrid,ticks,white
sns.set_palette('dark')
sns.boxplot(data_iris,orient='h')
plt.show()
```
