**Set: is an unordered collection of unique items,mutable.**

```python
our_set={'fan','ac','washing machine','cooler','ac'}
```

```python
print(our_set)
```

```python
#add(): to add the item in set
our_set.add('laptop')
print('Set after adding one more item')
for item in our_set:
    print(item)

```

```python
our_set.remove('cooler')
print('Set after removing one item')
for item in our_set:
  print(item)
```

```python
#union,intersection, difference
set_one={10,20,30,40,50}
set_two={5,10,15,20,25,30}
set_union=set_one.union(set_two)
print('Union of both sets')
print(set_union)
```

```python
set_intersection=set_one.intersection(set_two)
print('Intersection of both sets')
print(set_intersection)
```

```python
set_difference=set_one.difference(set_two)
print('Difference of both sets')
print(set_difference)
```

```python
set_difference=set_two.difference(set_one)
print('Difference of both sets')
print(set_difference)
```

```python
course_one={'Python','Power BI','Excel','SQL','Time Management','Soft Skill'}
course_two={'Python','Excel','SQL','Soft Skill','Tableau','TensorFlow','Azure','GitHub'}
print(course_one)
print(course_two)
```

```python
common_courses=course_one.intersection(course_two)
print('Common Skills')
for course in common_courses:
  print(course)
```

```python
course_union=course_one.union(course_two)
print('All Skills covered by both Training')
for course in course_union:
  print(course)
```

```python
course_diff_one=course_one.difference(course_two)
print('Skills Covered by Course_one Only')
for course in course_diff_one:
  print(course)
```

```python
course_diff_two=course_two.difference(course_one)
print('Skills Covered by course_two Only')
for course in course_diff_two:
  print(course)
```

Dictionary:is a mutable collection of key value pairs.
keys are unique

```python
players={'name':'Sachin','age':45,'country':'India'}
print(players)

```

```python
print('All Keys')
for k in players:
  print(k)
```

```python
print('All Keys')
for k in players.keys():
  print(k)
```

```python
print('All Values')
for v in players.values():
  print(v)
```

```python
print('Key -> Value')
for k in players:
  print(f'{k} ->{players[k]}')
```

```python
players['Type']='All-Rounder'
print(players)
```

```python
players['MatchFee']=5000
print(players)
```
