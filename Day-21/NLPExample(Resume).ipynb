```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
resumes=[
    'Python Developer  with experience in Deep Learning'
    'Experienced software engineer with expertise in web development and SQL Server',
    'Java Developer skilled in Spring Boot , SQL and Microservices',
    'Python Developer with experience in data analysis and machine learning',
     'Dot net Developer skilled in asp.net ,SQL , Microservices',
    'Dat Scientist with Python, Machine Learning ,NLP and Pandas',
    'Front End Developer with HTML, CSS-3 JavaScript and React'


]
```

```python
job_descriptions=[
    'Looking for Python Developer with experience in data analysis and machine learning',
]
```

```python
documents=resumes+job_descriptions
documents
```

```python
vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(documents)
#
```

```python
#compare resumes with job description
cosine_similarities=cosine_similarity(tfidf_matrix[:len(resumes)],tfidf_matrix[len(resumes):])
#cosine_similarities=cosine_similarity(tfidf_matrix[:-1],tfidf_matrix[-1])
```

```python
#Display score
for i,similarity_score in enumerate(cosine_similarities):
    print(f'Resume {i+1} Similarity Score: {similarity_score[0]:.2f}')
```
