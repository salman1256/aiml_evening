##Matplotlib: is a python library.
* It is used to create static , animated and interactive visualizations plot.
 * To Create plot we use matplotlib.pyplot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(
    {
        'Student Name':['Sam','Ravi','Amit','Vijay'],
        'Student Marks':[60,70,50,90]
    }
)
```

```python
print(df)
```

*Bar Chart*

```python

x=df['Student Name']
y=df['Student Marks']
plt.bar(x,y)
plt.title('Student Marks Bar Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.show()
```

```python
#Monthly Sales Data
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales=[50,100,170,190,210,230,250,270,290,310,330,350]
```

```python
x=range(len(months))
plt.xlim(1,12)
plt.ylim(50,400)
plt.plot(x,sales,color='magenta',linestyle='--',linewidth=2)
plt.xticks(x,months)
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
#
```

```python
x=range(len(months))
plt.xlim(1,12)
plt.ylim(0,350)
plt.plot(x,sales,color='magenta',linestyle='-',linewidth=2) #solid line
plt.xticks(x,months)
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.show()
```

```python
x=range(len(months))
plt.xlim(1,12)
plt.ylim(0,350)
plt.plot(x,sales,color='magenta',linestyle='-.',linewidth=2) #dashed line
plt.xticks(x,months)
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.show()
```

```python
x=range(len(months))
plt.xlim(1,12)
plt.ylim(50,350)
plt.plot(x,sales,color='magenta',linestyle=':',linewidth=2) #dotted line
plt.xticks(x,months)
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales')

plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sales_2024=[120,150,170,190,210,230,250,270,290,310,330,350]
sales_2025=[130,140,190,190,210,230,250,370,390,310,330,350]
sales_2026=[150,140,200,220,200,230,250,370,390,310,330,350]
```

**Bar Plot**: used to show the bar chart for categorical data usually.
on x-axis we show category and on y axis we show values.


```python
procuts=['Laptop','IPad','Ear Phone','Keyboard']
weekly_sales=[400,250,600,100]
```

```python
plt.bar(procuts,weekly_sales)
plt.title('Weekly Sales Report')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
#
```

```python
plt.figure(figsize=(6,4))
plt.bar(procuts,weekly_sales)
plt.title('Weekly Sales Report')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True,axis='y')
plt.tight_layout( )
plt.show()
#
```

```python
bar_colors=['magenta','skyblue','green','brown']
plt.figure(figsize=(6,4))
plt.bar(procuts,weekly_sales,color=bar_colors ,width=0.5, edgecolor='black')
plt.title('Weekly Sales Report')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True,axis='y')
plt.tight_layout( )
plt.show()
```

```python
products_sales={'Laptop':750,'Phone':650,'Watch':500, 'Tablets':600,'Head Phones':800,'Touch Screen':200}
products=list(products_sales.keys())
sales=list(products_sales.values())
print(products,'',len(products))
print(sales,len(sales))
```

```python
bar_colors=['magenta','skyblue','green','brown','yellow']
plt.figure(figsize=(6,4))
plt.bar(products,sales,color=bar_colors ,width=0.5, edgecolor='black')
plt.title('Weekly Sales Report')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True,axis='y')
plt.tight_layout( )
plt.show()
```
