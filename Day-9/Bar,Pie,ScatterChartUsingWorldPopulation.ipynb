```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/population_by_country_2020.csv')
```

```python
df.head(10)
```

```python
df_top=df.sort_values(by='Population (2020)',ascending=False).head(10)
```

```python
df_top
```

```python
countries=df_top['Country (or dependency)']
print(countries)
```

```python
population=df_top['Population (2020)']
print(population)
```

```python
bar_colors=plt.cm.Paired(np.arange(len(countries)))
print(bar_colors)
```

```python
plt.figure(figsize=(6,4))
plt.bar(countries,population,color=bar_colors)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries with the Highest Population in 2020')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

**Create a scatter plot using following data**
* height_cm=[150,155,160,165,170,175,180,185]
* weight_kg=[50,53,56,59,62,66,68,72]
* age_year=[20,22,25,28,30,35,38,40]
---

* Plot for persons height vs Weight
* Colored by Age

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,56,59,62,66,68,72]
age_year=[20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(6,4))
plt.scatter(height_cm,weight_kg,c=age_year,cmap='plasma',edgecolors='black',s=100)
plt.colorbar(label='Age')
plt.grid(color='lightblue',linestyle='--',linewidth=0.7,axis='x')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Height vs Weight')
plt.show()
```

**PIE CHART**
## A pie chart is a circular chart used to show a whole divide into parts.
* pie() functions to create pie plot in matplotlib

```python
departments=['HR','Web Development','App-Development','Agile-Scrum','Testing']
expenses=[30000,10000,30000,15000,15000]
```

```python
print(departments)
print(expenses)
```

```python
plt.figure(figsize=(5,5))
plt.title('Departments Expenses')
plt.pie(expenses,labels=departments,autopct='%1.0f%%',startangle=90,explode=(0.1,0.05,0.1,0.05,0.05),shadow=True)
plt.show()
```

```python
pie_Ccolors=plt.cm.tab20(np.arange(len(departments)))
pie_explode=((0.1,0.05,0.1,0.05,0.05))
plt.figure(figsize=(5,5))
plt.title('Departments Expenses')
plt.pie(expenses,labels=departments,autopct='%1.0f%%',startangle=90,explode=pie_explode,shadow=True,colors=pie_Ccolors)
plt.show()
```
