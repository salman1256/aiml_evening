```python
marks_per=int(input('enter your marks percentage without % sign: '))
grade='F'
if marks_per>90:
    grade='A'
elif marks_per>80:
  grade='B'
elif marks_per>70:
  grade='C'
elif marks_per>60:
  grade='D'

print(f'Marks: {marks_per} % Grade: {grade}')


```
