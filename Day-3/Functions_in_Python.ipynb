```python
#return: keyword used to return a value from function,Function can return the value back to calling enviroment
def avg_num(num1,num2):
  average=(num1+num2)/2
  return average
```

```python
fnum=int(input('Enter first number'))
snum=int(input('Enter second number'))
result=avg_num(fnum,snum)
print(f'The average of {fnum},{snum} numbers is {result}')
```

```python
res=avg_num(10,20)
print(res)
```

```python
#Variable Length Arguments
def display(*nums):
  print('\n All Numbers passed by you are:')
  for(i) in nums:
      print(i,end="\t")
```

```python
display(12,45)
display(10,5,15,20,45)
display(100,200,300)
```

## Write a function in python to calculate the cube of a number entered by the user.

* Take the number from user and display following details:
* for example
Number : 5  Cube : 125

```python
def cube_num(num):
  return num*num*num
```

```python
num=int(input('Enter a number to find out cube: \t'))

print('Cube of',num,'is',cube_num(num))
#
```

```python
#Recursive functions
def factorial(num):
  if num==1:
    return 1
  else:
    return num*factorial(num-1)

    #5: 5*4*3*2*1
```

```python
number=int(input('Enter Number to find out factorial: \t'))
print(f'Factorial of {number} is = \t',factorial(number) )
```

```python
def gen_table(number):
  for i in range(1,11):
    print(f"{number} x {i} = {number*i}")
```

```python
number=int(input('Enter number to find out table of number'))
gen_table(number)
```

```python
choice='y'
while(choice=='y'):
  number=int(input('Enter number to find out table of number: \t'))
  print(f'Table of number {number} as follows')
  gen_table(number)
  choice=input('Do you want to continue y/n')
```

```python
#lambda functions or Anonymous functions or one liner functions
cube=lambda num:num*num*num
print(cube(3))
```

```python
add=lambda num1,num2:num1+num2
print(add(10,20))
```
