**Exception Handling:**

```python
def addition(fnum,snum):
  result=fnum+snum
  return result
```

```python
try:
  num1=int(input("Enter first number: "))
  num2=int(input("Enter second number: "))
  print(f'Result after adding {num1} and {num2} =\t {addition(num1,num2)}')
except:
  print('Error Occured')
finally:
  print('End of Program!!')



```

```python
def div(fnum,snum):
  result=fnum/snum
  return result
```

```python
try:
  num=int(input('Enter Numerator: '))
  den=int(input('Enter Denominator: '))
  print(f'Resut after dividing {num} by {den} =\t',div(num,den))
except ZeroDivisionError as de:
  print('Error: Denominator cannot be zero: ',de)
except ValueError as ve:
  print('ValueError ',ve)
except Exception as e:
  print('Error: ',e)
finally:
  print('End of Program!!')
```
